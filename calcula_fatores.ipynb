```python
# Função que calcula os fatores

# Função que calcula fator da tabela financeira
def c_f(tipo, i, n):

    i = i/100

    if tipo == 'FP':
        fator = ((1 + i)**n)

    elif tipo == 'PF':
        fator = 1/((1+i)**n)

    elif tipo == 'FA':
        fator = ((1+i)**n - 1)/i
        
    elif tipo == 'AF':
        fator = i/((1+i)**n - 1)

    elif tipo == 'PA':
        fator = (((1+i)**n) - 1)/(i*(1+i)**n)

    elif tipo == 'AP':
        fator = (i*((1+i)**n)) / ((1+i)**n - 1)

    print(f"fator {tipo} = {fator}")

    return fator
    
    
```

### Exemplo de Aplicação

Um investimento no setor elétrico tem custo inicial de R$ 100.000,00 na data atual. Promete ao seu investidor um retorno anual de R$ 20.000,00 durante os próximos 10 anos. Sendo a TMA de 10% a.a., pede-se determinar o VPL do investimento e se ele deverá ser realizado.

```python
vlp = - 100000 + 20000*c_f("PF", 10, 10)


print(vlp)
```

Solução: com um VLP = 22891.342114093713

Conclusão: O investimento vale a pena.
